```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['image.origin'] = 'lower'
matplotlib.rcParams['figure.figsize'] = (5,4)
```

```python
import stpsf
```

```python
nc = stpsf.NIRCam()
```

```python
nc.filter='F444W'
# psf444 = nc.calc_psf(fov_arcsec=2,oversample=4, )
psf444 = nc.calc_psf(fov_pixels=20,oversample=4, )
stpsf.display_psf(psf444, ext=3)
```

```python
# display the PSF and plot the encircled energy
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)

stpsf.display_psf(psf444, colorbar_orientation='horizontal')
psf444.writeto('nircam_F444W.fits', overwrite=True)

axis2 = plt.subplot(1,2,2)
stpsf.display_ee(psf444, ax=axis2)

plt.savefig('fig_example_plot_nircam_f444w.png')
```
